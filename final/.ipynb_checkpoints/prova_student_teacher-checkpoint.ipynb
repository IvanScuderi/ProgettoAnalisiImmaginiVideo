### Funzioni di utilità

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def test(model, topk=10, activation=True):
    print('Start testing the model')
    predictions = []
    y_true = []
    model.eval()
    with torch.no_grad():
        for x, y in testDataLoader:
            x, y = x.to(device), y.to(device)
            y_pred = model(x)
            
            if activation:
                y_pred = torch.sigmoid(y_pred)
            
            _, idx = y_pred.topk(topk, dim=1)
            y_pred = torch.zeros_like(y_pred)
            
            y_pred.scatter_(1, idx, 1)
            predictions.append(y_pred.cpu())

            y_true.append(y.cpu())
    y_true, predictions = torch.cat(y_true, axis=0), torch.cat(predictions, axis=0)
    pr,re,f1,_ = precision_recall_fscore_support(y_true, predictions, average='micro')
    print("Test END: -- m_precision = ",np.round(pr,2), " -- m_recall = ",np.round(re,2)," -- round m_f1 = ",np.round(f1,2),
         " -- m_f1 = ",f1)
```

### Generazione Modello Teacher

Definisco alcuni metodi che mi serviranno per il salvataggio/caricamento del modello/checkpoint.

```python
CHECKPOINT_TEACHER = os.path.join(DRIVE_TEACHER_MODEL, 'model.checkpoint')
MODELFILE_TEACHER = os.path.join(DRIVE_TEACHER_MODEL, 'model.pth')

CHECKPOINT_STUDENT = os.path.join(DRIVE_STUDENT_MODEL, 'model.checkpoint')
MODELFILE_STUDENT = os.path.join(DRIVE_STUDENT_MODEL, 'model.pth')

def save_checkpoint(epoch, model, optimizer, loss, teacher=True):
    if teacher:
        torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': loss,
                }, CHECKPOINT_TEACHER)
    else:
        torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': loss,
                }, CHECKPOINT_STUDENT)

def load_checkpoint(model, optimizer, teacher=True):
    if teacher:
        if not os.path.exists(CHECKPOINT_TEACHER):
            return None, None
        checkpoint = torch.load(CHECKPOINT_TEACHER)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        epoch = checkpoint['epoch']
        loss = checkpoint['loss']

        return epoch, loss
    else:
        if not os.path.exists(CHECKPOINT_STUDENT):
            return None, None
        checkpoint = torch.load(CHECKPOINT_STUDENT)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        epoch = checkpoint['epoch']
        loss = checkpoint['loss']

        return epoch, loss

def save_model(model, teacher=True):
    if teacher:
        torch.save(model.state_dict(), MODELFILE_TEACHER)
        print('Salvataggio modello Teacher avvenuto con successo!')
    else:
        torch.save(model.state_dict(), MODELFILE_STUDENT)
        print('Salvataggio modello Student avvenuto con successo!')

def load_model(model, teacher=True):
    if teacher:
        if os.path.exists(MODELFILE_TEACHER):
            model.load_state_dict(torch.load(MODELFILE_TEACHER, map_location=torch.device('cpu')))
            print("Caricamento modello Teacher avvenuto con successo!")
    else:
        if os.path.exists(MODELFILE_STUDENT):
            model.load_state_dict(torch.load(MODELFILE_STUDENT))
            print("Caricamento modello Student avvenuto con successo!")
```

```python
net = models.vgg19(pretrained = True)

class Identity(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, x):
        return x

net.fc2 = Identity()

fully_connected = nn.Sequential(
    nn.Conv1d(20,20,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(20),
    nn.Conv1d(20,32,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(32),
    nn.MaxPool1d(2),
    nn.Dropout(0.5),
    nn.Conv1d(32,64,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.Conv1d(64,64,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.MaxPool1d(2),
    nn.Dropout(0.5),
    nn.Flatten(),
    nn.Linear(1024,512),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(512,85),
    nn.Sigmoid()
)

class Teacher(nn.Module):
    def __init__(self,backbone,fully_connected):
      super().__init__()
      self.backbone = backbone
      self.fully_connected = fully_connected

    def forward(self,x):
        vector = torch.zeros((x.size(0),limit,1000))
        i = 0
        for trailer in x:
            output = self.backbone(trailer) 
            vector[i] = output
            i +=1
        result = self.fully_connected(vector.to(device))
        return result


teacher = Teacher(net,fully_connected)

for params in teacher.parameters():
  params.requires_grad = False

print('Numero parametri modello Teacher: ',count_parameters(teacher))

teacher
```

### Generazione Modello Student

```python
net_s = models.vgg19(pretrained = True)

for params in net_s.parameters():
  params.requires_grad = False

print('Numero parametri backbone: ',count_parameters(net_s))

net_s
```

```python
fully_connected_s = nn.Sequential(
    nn.Conv1d(20,20,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(20),
    nn.Conv1d(20,32,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(32),
    nn.MaxPool1d(2),
    nn.Dropout(0.5),
    nn.Conv1d(32,64,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.Conv1d(64,64,kernel_size=(3),padding=1,stride=2),
    nn.ReLU(),
    nn.BatchNorm1d(64),
    nn.MaxPool1d(2),
    nn.Dropout(0.5),
    nn.Flatten(),
    nn.Linear(1024,512),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(512,85)
)

print('Numero parametri fully connected layers: ',count_parameters(fully_connected_s))

fully_connected_s
```

```python
net_s.fc2 = Identity()
```

```python
class Student(nn.Module):
    def __init__(self,backbone,fully_connected):
      super().__init__()
      self.backbone = backbone
      self.fully_connected = fully_connected

    def forward(self,x):
        vector = torch.zeros((x.size(0),limit,1000))
        i = 0
        for trailer in x:
            output = self.backbone(trailer) 
            vector[i] = output
            i +=1
        result = self.fully_connected(vector.to(device))
        return result
```

```python
model = Student(net_s,fully_connected_s)
model = model.to(device)

print('Numero parametri model: ',count_parameters(model))

model
```

Vado a definire i pesi che userà la loss in modo tale da dare maggiore importanza agli errori commessi sulle classi poco numerose.

```python
weight = [1/math.sqrt(elem) for elem in count]

for i in range(len(weight)):
    if weight[i] == 1:
        weight[i] = 0

weight = torch.FloatTensor(weight)
weight = weight.to(device)

weight
```

```python
lr = 1e-3
criterion_lab = nn.BCEWithLogitsLoss(weight=weight)
criterion_ps = nn.BCEWithLogitsLoss(weight=weight)
optimizer = torch.optim.Adam(model.parameters(),lr=lr)
epoca = 0
```

```python
load_model(model, False)
```

Siccome il modello Teacher produce come risultato un tensore di valori Float che corrispondono alle probabilità di ogni label per il determinato trailer, vado a trasformare questo in un tensore binario associando 1 alle topk=5 label più probabili secondo il Teacher. Si usa il valore 5 siccome precedentemente abbiamo notato che la maggior parte dei trailer hanno alpiù 5 label assegnate.

```python
teacher = teacher.to(device)

epoca, error = load_checkpoint(model,optimizer, False)

epochs = 16
train_losses = []
test_losses = []
last_epoch = epoca

topk_training = 3

print_every = 20

noyse =  torchvision.transforms.Compose([
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.RandomVerticalFlip(),
    torchvision.transforms.RandomRotation(degrees=(-30, 30)),
])

try:
    for epoch in range(last_epoch or 0, epochs):
        print('Start epoch', epoch+1)
        model.train()
        running_loss = 0
        steps = 0
        for (x, y), x_ps in zip(trainingDataLoader, pseudoLabeledDataLoader):
            steps += 1
            
            #x = x.view(16,20,3,224,224)
            x, y = x.to(device), y.to(device)
            x_ps = x_ps.to(device)
            
            teacher.eval()
            with torch.no_grad():
                y_ps = teacher(x_ps)
            
            y_ps = y_ps.cpu()
            top = y_ps.topk(topk_training, dim=1)
            
            ones = torch.ones(y_ps.size())
            zeros = torch.zeros(y_ps.size())
            zeros = zeros.scatter_(1,top.indices,ones)
            
            y_ps = zeros.to(device)
            
            #applico il noyse ad x_ps
            x_ps = x_ps.cpu()
            for i in range(batch_size):
                for j in range(limit):
                    x_ps[i,j] = noyse(x_ps[i,j])
            x_ps = x_ps.to(device)
            
            optimizer.zero_grad()
            
            logit = model(x)
            logit_ps = model(x_ps)

            loss_lab = criterion_lab(logit, y)
            loss_ps = criterion_ps(logit_ps, y_ps)*1.4
            
            loss = loss_lab + loss_ps
            loss.backward()

            optimizer.step()
            running_loss += loss.item()
            
            if steps % print_every == 0:
                print(f"epoch {epoch+1}/{epochs} "
                      f"train loss: {running_loss/steps:.6f} ")

                save_checkpoint(epoch, model, optimizer, loss, False)

        train_losses.append(running_loss/steps)
        print(f"END Epoch {epoch+1}/{epochs} "
              f"Train loss: {running_loss/steps:.6f} ")
        save_model(model, False)
        if (epoch+1)%1 == 0:
            # effettuo il test ogni due epoche
            test_losses.append(test(model))
except KeyboardInterrupt: 
    print('Exiting from training early')

save_model(model, False)
```

Vado ad effettuare il plot dell'errore durante la fase di training sui dati labeled.
